### Text Cleaning

##### Removing the Symbols and Numbers

```python
def clean_func(Text):
  Text = re.sub(r'[\([{})\]!@#$,"%^*?:;~`0-9]', ' ', Text)   # removing the symbols and numbers
  Text = Text.lower()   # converting the text to lower case
  Text = re.sub('#\S+', '', Text)  # remove hashtags

  return Text 

search['Module clean'] = search['tags'].apply(lambda x:clean_func(x))
```

##### Lemmatization

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_words(text):
    words = text.split()
    words = [lemmatizer.lemmatize(word,pos='v') for word in words]
    return (words)

search['lemmatized'] = search['Module clean'].apply(lemmatize_words)
```

##### Stop Word Removal

```python
stop = stopwords.words('english')
search['Module Tags'] = search['lemmatized'].apply(lambda x: ' '.join([word for word in x if word not in (stop)]))
search['Module Tags'] = search['Module Tags'].str.replace('[-_]',' ')
search['Module Tags'] = search['Module Tags'].str.replace('  ',' ')
search.head()
```

### Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend_course_st1(inp):
  inp = [inp]
  inp = st1.encode(inp)
  inp = inp.reshape(1,-1)
  x = s.shape[0]
  sim=[]
  for i in range(0,x):
    n = s['st_vector1'][i]
    
    sim.append(cosine_similarity(n,inp))
  scores = list(enumerate(sim))
  scores = sorted(scores, key=lambda n: n[1],reverse=True)
  scores = scores[1:6]
  course = [j[0] for j in scores]
  modules = s['Module Name'].iloc[course]
  m_id = s['Module Id'].iloc[course]
  result = [j[1] for j in scores]
  r = np.array(result)
  f = r.flatten()
  new = pd.DataFrame({'Module ID' : m_id, 'Module Name' : modules, 'Similarity Scores' : f }).reset_index()
  new.drop(['index'], axis=1, inplace = True)
  return new
```
